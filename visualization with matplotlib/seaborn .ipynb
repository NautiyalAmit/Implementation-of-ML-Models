```python
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
sns.set()
```

```python
# Create some data
rng = np.random.RandomState(0)
x = np.linspace(0, 10, 500)
y = np.cumsum(rng.randn(500, 6), 0)
```

```python
# Plot the data with Matplotlib defaults
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left');
```

```python
##Histograms, KDE, and densities
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])
for col in 'xy':
    plt.hist(data[col], density=True, alpha=0.5)
```

```python
#Kernel density estimates for visualizing distributions
for col in 'xy':
    sns.kdeplot(data[col], shade=True)
```

```python
#Kernel density and histograms plotted together
sns.distplot(data['x'])
sns.distplot(data['y']);
```

```python
#A two-dimensional kernel density plot
sns.kdeplot(data);
```

```python
#A joint distribution plot with a two-dimensional kernel density estimate
with sns.axes_style('white'):
    sns.jointplot("x", "y", data, kind='kde');
```

```python
#A joint distribution plot with a hexagonal bin representat
with sns.axes_style('white'):
    sns.jointplot("x", "y", data, kind='hex')
```
