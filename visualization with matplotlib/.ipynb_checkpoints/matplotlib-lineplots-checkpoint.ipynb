```python
## import 
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

```

```python
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))

plt.show()

```

```python
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x),'--')
plt.plot(x,np.cos(x),'-')

plt.show()
```

```python
##subplots 
plt.subplot(2,1,1)
plt.plot(x,np.sin(x))

plt.subplot(2,1,2)
plt.plot(x,np.cos(x))

plt.show()
```

```python
##object oriented interfafce


fig, ax=plt.subplots(2)

#call 
ax[0].plot(x,np.sin(x))
ax[1].plot(x,np.cos(x))
```

```python
plt.style.use('seaborn-whitegrid')
#draw figurew and axes 
fig=plt.figure()
ax=plt.axes()
```

```python
plt.style.use('seaborn-whitegrid')
#draw figurew and axes 
fig=plt.figure()
ax=plt.axes()
ax.plot(x,np.sin(x),color='blue',linestyle='solid')
ax.plot(x,np.cos(x),color='g',linestyle='dotted')

```

```python
plt.style.use('seaborn-whitegrid')
#draw figurew and axes 
fig=plt.figure()
ax=plt.axes()
#reshaping axis
ax.plot(x,np.sin(x))
plt.axis([-1,10,-1,20])
```

```python
plt.style.use('seaborn-whitegrid')
#draw figurew and axes 
fig=plt.figure()
ax=plt.axes()
#reshaping axis
ax.plot(x,np.sin(x))
plt.axis('equal');
```

```python
#labeloing the plots 
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.legend();
```

```python
#plot all the funtions simultaneously 
#basis
plt.style.use('seaborn-whitegrid')
fig=plt.figure()
ax=plt.axes()
#x,y define
ax.plot(x,np.sin(x))



ax.set(xlim=(0, 10), ylim=(-2, 2),
       xlabel='x', ylabel='sin(x)',
       title='A Simple Plot');
```
